```python
from sklearn.preprocessing import LabelEncoder
```

```python
df['target_encoded'] = df['diagnosis'].map({'M': 0, 'B': 1})
```

```python
df.drop("diagnosis", axis=1,inplace=True)
```

```python
df
```

```python
x = df.iloc[:, :30]
```

```python
x
```

```python
y=df["target_encoded"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
y_test.shape
y_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#model=DecisionTreeClassifier()
model=RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_predicted=model.predict(X_test)
```

```python
y_predicted.shape
```

```python
from sklearn.metrics import classification_report
```

```python
y_test.shape
```

```python
y_predicted.shape
```

```python
print(classification_report(y_test, y_predicted))
```

```python

import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = pd.Series([10, 20, 5, 30, 25,848,515,435456,89451,5651,5646516,5645665,5465465,465456,498451,465165,498161])

# Create plot
plt.plot(data, marker='o')

# Find max value and its index
max_value = data.max()
max_index = data.idxmax()

# Annotate max point
plt.annotate(f'Max: {max_value}',
             xy=(max_index, max_value),              # point to annotate
             xytext=(max_index, max_value-5),      # text position
             arrowprops=dict(facecolor='red', shrink=0),
             font="Times New Roman",fontsize=10, color='black')

# Show plot
plt.title("Annotating Maximum Value",font="Times New Roman")
plt.xlabel("Index",font="Times New Roman")
plt.ylabel("Value",font="Times New Roman")
plt.grid(True)
plt.show()

```
